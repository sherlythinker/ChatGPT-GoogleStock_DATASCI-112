Import News Data and Perform Some Cleaning

```python
news_df = pd.read_csv("/content/Feb19-Mar18.csv(1)")
news_df["content"] = news_df["content"].fillna("0")
news_df["description"] = news_df["description"].fillna("0")
news_df["title"] = news_df["title"].fillna("0")
news_df
```

Import SentimentIntensityAnalyzer and Run on each article's Title, Description, and Content

```python
import nltk

# Download the required nltk resources
nltk.download('vader_lexicon')

# Load the sentiment analyzer
from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

lst = []
# Define a function to calculate the sentiment of each row
for index, row in news_df.iterrows():
  lst.append(sia.polarity_scores(row['content'])) 

df_new = pd.DataFrame(lst)
news_df = pd.concat([news_df, df_new], axis=1)
```

```python
# Perform some cleaning, correct column names
news_df = news_df.rename(columns={"neg": "neg_content", "neu": "neu_content", "pos": "pos_content", "compound": "compound_content"})
news_df
```

```python
# Load the sentiment analyzer
from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
news_df['title'] = news_df['title'].fillna("0")
lst = []
# Define a function to calculate the sentiment of each row
for index, row in news_df.iterrows():
  lst.append(sia.polarity_scores(row['title'])) 

df_new = pd.DataFrame(lst)
news_df = pd.concat([news_df, df_new], axis=1)

# Perform some cleaning, correct column names
news_df = news_df.rename(columns={"neg": "neg_title", "neu": "neu_title", "pos": "pos_title", "compound": "compound_title"})
news_df
```

```python
from numpy import newaxis
# Load the sentiment analyzer
from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
news_df['description'] = news_df['description'].fillna("0")
lst = []
news_df
# Define a function to calculate the sentiment of each row
for index, row in news_df.iterrows():
  lst.append(sia.polarity_scores(row['description'])) 

df_new = pd.DataFrame(lst)
news_df = pd.concat([news_df, df_new], axis=1)

# Perform some cleaning, correct column names
news_df = news_df.rename(columns={"neg": "neg_description", "neu": "neu_description", "pos": "pos_description", "compound": "compound_description"})
news_df
```
